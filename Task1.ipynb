```python
from matplotlib import pyplot as plt
%matplotlib inline
import random

random_array = [random.randint(0, 100) for _ in range(20)]
print(random_array)
```

```python
def find_deepest_lake(heights):
    
    deepest_lake_heights = []
    deepest_lake_indexs = []
    max_deep = 0

    i=0
    while i < len(heights) - 1:

        #find first coast of lake
        if(heights[i]<= heights[i+1]):
            i+=1
            continue

        first_coast = i
        min_index = i + 1

        #find local deep of lake
        while i + 1 < len(heights) and heights[i] > heights[i+1]:
            if heights[i+1] < heights[min_index]:
                min_index = i + 1
            i += 1
        
        second_coast = i
        
        #find second coast of lake
        while i + 1 < len(heights)  and heights[i] <= heights[i+1]:
            if heights[i+1] > heights[second_coast]:
                second_coast = i + 1
            i += 1

        #find deepest lake
        if first_coast < second_coast and heights[second_coast] > heights[min_index]:
            local_min = min(heights[first_coast], heights[second_coast])
            local_deep = local_min - heights[min_index]

            if local_deep > max_deep:
                max_deep = local_deep

                deepest_lake_indexs = list(range(first_coast, second_coast+1))
                deepest_lake_heights = [heights[x] for x in deepest_lake_indexs]

        i = second_coast

    return deepest_lake_heights, deepest_lake_indexs
```

```python
deepest_lake, lake_indexes = find_deepest_lake(random_array)

plt.plot(random_array, linestyle='-')
plt.plot(lake_indexes, deepest_lake, '')
plt.show()
```
